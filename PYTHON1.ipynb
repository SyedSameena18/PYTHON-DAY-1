```python
import random
p1=input("Enter player1: ")
p2=input("Enter player2: ")
s1=10
s2=10
d1=random.randint(1,10)
d2=random.randint(1,10)

print("-----PLAYER1 TURN-----")
while True:
    g1=input("Enter ur guess: ")
    if not g1.isdigit():
        print("Please enter a valid number")
        continue
    g1=int(g1)
    if g1==d1:
        print("Correct guess!")
        break
    s1=s1-1
    if s1<=0:
        print("Player 1 ran out of guesses.")
        break

print("-----PLAYER2 TURN-----")
while True:
    g2=input("Enter ur guess: ")
    if not g2.isdigit():
        print("Please enter a valid number")
        continue
    g2=int(g2)
    if g2==d2:
        print("Correct guess!")
        break
    s2=s2-1
    if s2<=0:
        print("Player 2 ran out of guesses.")
        break

print("PLAYER1 score:",s1)
print("PLAYER2 score:",s2)

if s1>s2:
    print(p1,"wins")
elif s2>s1:
    print(p2,"wins")
else:
    print("Draw")
```
