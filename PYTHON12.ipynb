```python
class Graph:
    def __init__(self, v):
        self.adj = [[] for i in range(v)]

    def add_edge(self, a, b):
        self.adj[a].append(b)
        self.adj[b].append(a)

    def display(self):
        for i in range(len(self.adj)):
            print("Node {}:".format(i), end=" ")
            for j in self.adj[i]:
                print(j, end=" --> ")
            print()
a = Graph(6)
a.add_edge(0, 2)
a.add_edge(0, 3)
a.add_edge(1, 4)
a.add_edge(1, 3)
a.add_edge(3, 2)
a.add_edge(3, 3)
a.add_edge(5, 1)
a.display()
```
