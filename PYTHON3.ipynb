```python
# blinking of text
import time
import sys
h=3
m=40
s=0
while True:
   while True:
    try:
        sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
        sys.stdout.flush()
        time.sleep(1)
        s+=1
        if s==60:
            s=0
            m+=1
        if m==60:
            m=0
            h+=1
        if h==13:
            h=1
            m=0
            s=0
    except KeyboardInterrupt:
        print('\nStopped.')
        break
```
