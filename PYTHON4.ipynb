```python
names = ["raM","RAJ","Rose","JaCk","PaVi"]
marks=[[30,40,50],[66,77,55],[33,78,90],[87,97,57],[47,88,44]]
for i in range(5):
  p=sum(marks[i])//3
  if p>80:
    grade="S"
  elif p>=61:
    grade="A"
  elif p>=40 and p<60:
    grade="B"
  else:
    grade="Fail"
  print("Mr/Ms {} has scored {}% - {} Grade".format(i+1,names[i].capitalize(),p,grade))
```

```python
# Productbasedcompany
# Scheduling algorithm
import random
a=int(input("Enter num of teams: "))
teams=[]
print("Enter all team name: ")
for i in range(a):
    t=input("Enter all team name: ")
    teams.append(t)
m=int(input("Enter num of meet btw teams: "))
matches=[]
for i in range(0,a-1):
  for j in range(i+1,a):
    for k in range(m):
      matches.append((teams[i],teams[j]))
random.shuffle(matches)
pos=1
for i in matches:
  print("MATCH {}:{} VS {}".format(pos,i[0],i[1]))
  pos=pos+1
```
