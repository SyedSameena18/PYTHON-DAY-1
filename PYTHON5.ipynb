```python
# FLAMES
#CONVERTED str to list because its mutable
a=input("Enter boy name:")
b=input("Enter girl name:")
a1=list(a)
b1=list(b)
for i in range(len(a1)):
  for j in range(len(b1)):
    if a1[i]==b[j]:
      a1[i]='2'
      b1[j]='2'
print(a1)
print(b1)
cnt=0
for i in a1:
    if i!='2':
      cnt=cnt+1
for j in b1:
    if j!= '2':
      cnt=cnt+1
print(cnt)
index=0
res=list("FLAMES")
for i in range(5):
  index=(index+cnt)%len(res)
  res.pop(index)
print(res[0])
```
