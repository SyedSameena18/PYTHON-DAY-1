```python
def prime(n):
    if n<=1:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False
        return True
def sod(n):
    d=[int(i) for i in str(n)]
    return sum(d)
def every(n):
    for i in range(3):
        d=n%10
        n=n//10
        if not prime(d):
            return False
    return True

for i in range(100,1000):
    if(prime(i) and prime(sod(i)) and every(i)):
        print(i,end=" ")
```

```python
def cop(i,j):
    return gcd(i,j)==1
def gcd(a,b):
    while(b!=0):
        rem=a%b
        a=b
        b=rem
    return a

a = int(input("Enter any number: "))
for i in range(5,a):
    for j in range(4,i):
        for k in range(3,j):
            if ((j*j) +(k*k)==(i*i) and cop(i,j) and cop(i,k) and cop(j,k)):
                print(k,j,i)
```
