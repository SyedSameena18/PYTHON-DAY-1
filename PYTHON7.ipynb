```python
# A monkey to reach top of tower height x meters from the bottom.
#In the duration of first minute monkey will climb 'Y' metre.
#Next minute monkey will slips down 'Z' meter. Again in third minute it gains 'Y' meter then slips down 'Z' meter in next minute and it continues. Find total number of minutes to reach to top??
#Testcase 1:
#X: 30 meters, Y: 10 meters, Z: 5 meters
#output is 9 minutes
x=30
y=10
z=5
time=0
ht=0
i=1
while(ht<x):
  if i%2!=0:
    ht=ht+y
  else:
    ht=ht-z
  time=time+1
  i=i+1
print(time)
```

```python
#Leetcode 412
class Solution:
    def fizzBuzz(self,n:int):
        res=[]
        for i in range(1,n+1):
            if i%3==0 and i%5==0:
                res.append("FizzBuzz")
            elif i%3==0:
                res.append("Fizz")
            elif i%5==0:
                res.append("Buzz")
            else:
                res.append(str(i))
        return(res)
```
