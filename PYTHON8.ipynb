```python
def naresh(a):
  if a==1:
     return 1
  else:
     return a * naresh(a-1)

a=6
x=naresh(a)
print(x)
```

```python
def sameena(a):
  if a==1:
     return 1
  else:
     return a * sameena(a-1)

a=7
x=sameena(a)
print(x)
```

**OOPS**
variable-used to store or hold a value

4 types

1.Local vars: scope only in particular method

2.instance(global): each and every object hold unique value

3.class var(global): every obj holds same value

4.complete global: holds common value and used anywhere in the program

class-a container which contains methods and variables

Object-member of class

Methods and variables are displayed by mentioned using object name

```python
class abc:
  def set_dim(self,a,b):
    self.total=a+b #instance variable
  def display(self):
    print("Addition is",self.total)

a=abc()
b=abc()
a.set_dim(20,30) #local variable
b.set_dim(30,50) #local variable
a.display()
b.display()
```

```python
class abc:
  num=100 #class variable
  def set_dim(self,a,b):
    self.total=a+b+abc.num #instance variable
  def display(self):
    print("Addition is",self.total)

a=abc()
b=abc()
a.set_dim(20,30) #local variable
b.set_dim(30,50) #local variable
a.display()
b.display()
```

```python
class abc:
    num = 100  # class variable

    def set_dim(self, a, b):
        self.total = a + b + abc.num  # instance variable
        self.display()

    def display(self):
        print("Addition is", self.total)

a = abc()
b = abc()

a.set_dim(20, 30)
b.set_dim(30, 50)
```

```python
#finding prime factors
def sameena(a):
  if a==1:
    return
  i=2
  while(a%i!=0):
    i=i+1
  print(i,end=" ")
  sameena(a//i)
a=int(input("Enter any num:"))
sameena(a)
```

```python
class Rect:
    def __init__(self, a, b):
        self.area = a * b       # instance variable
    def display(self):
        print("Area is", self.area)

r = Rect(20, 30)
r2 = Rect(40, 67)
r.display()
r2.display()
```
