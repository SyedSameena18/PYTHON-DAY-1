```python
# int to list we use comprehension
a=234
b=[int(i) for i in str(a)]
print(b)
```

```python
# int to dict add element in dict using comprehension
a=3333322267
b={}
for i in str(a):
  if i not in b:
    b[i]=1
  else:
    b[i]-b[i]+1
print(b)



```

```python
# DICT TO LIST we use comprehension
b={'A':2,'B':7,'C':9}
a1=[i[0] for i in b.items()]
a2=[i[1] for i in b.items() if i[1]>2]
print(a1)
print(a2)
```

# functions:
1) infinite args - function can accept any number of arguments using *
2) default args - parameter is assigned a default value



```python
def abc(*args):
  sum=0
  for i in args:
    sum=sum+i
  print(sum)
abc(20,30)
abc(56,6,6,6)

```
